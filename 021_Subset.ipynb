```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# data_RAW = pd.read_csv('EmissiesVulwagens_202501.csv')
data_RAW = pd.read_csv('Data/VW3_20250124_1311_228.csv', delimiter=',')
```

```python
data_toclean = data_RAW.copy()
```

```python
# Renaming the first column
data_toclean.rename(columns={'Unnamed: 0': 'Timestamp'}, inplace=True)
data_toclean.head()
```

```python
data_toclean_dti = data_toclean.copy()
data_toclean_dti.info()
```

```python
data_toclean_dti['Timestamp'] = pd.to_datetime(data_toclean_dti['Timestamp'], errors='coerce')
data_toclean_dti.info()
```

```python
data_toclean_dti.head()
```

```python
# Setting the 'Timestamp' column as the DatetimeIndex
data_toclean_dti.set_index('Timestamp', inplace=True)
data_toclean_dti.head()
```

```python
data_toclean_dti.describe()
```

```python
# Condities bepalen zone vijzelen dP's obv #toeren vijzels
condition_vzl_start = (data_toclean_dti['Vijzel3'] > 1.0) | (data_toclean_dti['Vijzel4'] > 1.0)
condition_vzl_end = (data_toclean_dti['Vijzel3'] > (data_toclean_dti['Vijzel3'].max()- 3.0)) & (data_toclean_dti['Vijzel4'] > (data_toclean_dti['Vijzel4'].max()- 3.0))
condition_vzl_start, condition_vzl_end

# Zone calibratie dP's
zone_vzl_start = data_toclean_dti[condition_vzl_start].index[0]
zone_vzl_end = data_toclean_dti[condition_vzl_end].index[0]
zone_vzl_start, zone_vzl_end
```

```python
# Calculate the mean of values before the subset_timestamp
dP3_cal = data_toclean_dti[data_toclean_dti.index < zone_vzl_start]['dP3'].mean()
dP4_cal = data_toclean_dti[data_toclean_dti.index < zone_vzl_start]['dP4'].mean()
dP3_cal, dP4_cal
```

```python
data_toclean_dti['dP3'] = data_toclean_dti['dP3'] - dP3_cal
data_toclean_dti['dP4'] = data_toclean_dti['dP4'] - dP4_cal
data_toclean_dti.head()
```

```python
# Ovennummer bepalen
OvenNr = data_toclean_dti['OvenNr'].iloc[0]

# Figuur plotten
fig, ax1 = plt.subplots(figsize=(25, 10))

# Plotting the first axis
ax1.plot(
    data_toclean_dti.index,
    data_toclean_dti['dP3'].values,
    'b-',
    label='dP3',
)
ax1.plot(
    data_toclean_dti.index,
    data_toclean_dti['dP4'].values,
    'r-',
    label='dP4',
)
ax1.set_xlabel('date')
ax1.set_ylabel('mBar')

plt.title(f'dP3/4 - OvenNr: {OvenNr}')
plt.axvspan(zone_vzl_start, zone_vzl_end, color='yellow', alpha=0.1)
plt.axvspan(zone_vzl_start+pd.Timedelta(minutes=1), zone_vzl_end, color='green', alpha=0.1)
plt.axhline(y=0, color='k', linestyle='--', label='0 mBar')
plt.legend()
plt.show()
```
