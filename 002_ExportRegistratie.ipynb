## Basisdataset construeren obv RAW-data
1. Copy RAW-data
1. Selectie kolommen, eventueel hernoemen
1. Eventueel index aanpassen

```python
data_toclean = data_RAW.copy()
```

```python
# Fill missing values in the header with the previous value
data_toclean.columns= data_toclean.columns.fillna(method='ffill')
```

```python
# Convert MultiIndex columns to a DataFrame
columns_df = pd.DataFrame(data_toclean.columns.tolist())
# Fill missing values with the previous value
columns_df = columns_df.fillna(method='ffill')
# Convert the DataFrame back to a MultiIndex
data_toclean.columns = pd.MultiIndex.from_frame(columns_df)
# Display the updated headers
print(data_toclean.columns)
```

```python
first_header = data_toclean.columns.get_level_values(0).to_list()
first_header_filled = first_header.to_series().fillna(method='ffill').values
first_header_filled
```

```python
# Extract the first and second header rows
first_header = data_toclean.columns.get_level_values(0).tolist()
second_header = data_toclean.columns.get_level_values(1).tolist()
# Replace 'Unnamed' column names in the first header row
for i in range(1, len(first_header)):
 if 'Unnamed' in first_header[i]:
  first_header[i] = first_header[i - 1]
```

```python
first_header
```

```python
# Combine the updated first header row with the second header row
data_toclean.columns = pd.MultiIndex.from_arrays([first_header, second_header])
# Display the updated headers
print(data_toclean.columns)
```

```python
data_toclean.columns = ['_'.join(col).strip() for col in data_toclean.columns.values]
```

```python
# Display the updated headers
print(data_toclean.columns)
```
